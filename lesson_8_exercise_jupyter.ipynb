## !!!Exercise: 
### write a for loop that looks at each element of the following loop
### and assigns a value to a new list stating if the temperature is cold, mild, warm or hot

```python
temp_list = [100, 43, 10, 6, 68, 90, 24, 34]
new_list = []

for i in range(len(temp_list)):


    if temp_list[i] < 40:
        new_list.append("cold")
    elif temp_list[i] > 40 and temp_list[i] < 60:
        new_list.append("mild")
    elif temp_list[i] > 60 and temp_list[i] < 80:
        new_list.append("warm")
    else: 
        new_list.append("hot")


#print the full final list to check your work

print(new_list)
```

## while loops
### you can do the same things with while loops as for loops

## fix it

### while loops are mostly used when there's a natural stopping condition
### make a loop that counts to 100 by 10s

```python
count = 0
while count != 100:
    count += 10
print(count)
```

## Exercise: 
### create a while loop that counts the number of random draws needed to 
### draw the number 53.

```python
#this code will create a random number between 1 and 100
import random
draw = 0
num = 0
while draw != 53:
    draw = random.randint(1,100)
    num += 1
print(num)
```
