!putting it all together


```python

with open(r'C:\Users\noah.ross\OneDrive - West Point\Desktop\AY26-2\SE370/sample.csv', 'r') as file:
    lines = file.readlines()
    data = []
    for line in lines:
        data.append(line.strip.split(','))

    print(data)
```
